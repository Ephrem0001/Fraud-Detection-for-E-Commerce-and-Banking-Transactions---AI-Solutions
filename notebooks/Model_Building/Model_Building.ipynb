```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score, accuracy_score
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.preprocessing import StandardScaler
import mlflow
import mlflow.sklearn
pd.set_option('display.max_columns', None)
```

```python
file_path = r"C:/Users/ephre/Documents/Fraud-Detection-for-E-Commerce-and-Banking-Transactions---AI-Solutions/Data/Cleaned_Data/combined_data.csv"
combined_data = pd.read_csv(file_path)
combined_data.head()
```

```python

data = combined_data.drop('class_fraud', axis=1)
target = combined_data['class_fraud']

# Train-test split (80% train, 20% test)
X_train, X_test, y_train, y_test = train_test_split(data, target, test_size=0.2, random_state=42)


```

```python
models = {
    'Logistic Regression': LogisticRegression(max_iter=500),
    'Decision Tree': DecisionTreeClassifier(),
    'Random Forest': RandomForestClassifier(),
    'Gradient Boosting': GradientBoostingClassifier(),
    'MLP': MLPClassifier(max_iter=1000)
}
```

```python
# Assuming the target has continuous values, convert it to binary (0 or 1)
threshold = 0.5  # Define a threshold for binary classification
y_train = (y_train >= threshold).astype(int)  # Convert continuous to binary
y_test = (y_test >= threshold).astype(int)
```

```python
import mlflow.models.signature as signature

# Example input for logging
input_example = X_test[:1]

# Automatically infer the model's signature
model_signature = signature.infer_signature(X_train, model.predict(X_train))

for model_name, model in models.items():
    with mlflow.start_run(run_name=model_name):
        # Train the model
        model.fit(X_train, y_train)
        
        # Predict on test data
        y_pred = model.predict(X_test)
        
        # Calculate metrics
        acc = accuracy_score(y_test, y_pred)
        roc_auc = roc_auc_score(y_test, model.predict_proba(X_test)[:, 1])
        conf_matrix = confusion_matrix(y_test, y_pred)
        
        # Log parameters and metrics
        mlflow.log_param("model_name", model_name)
        mlflow.log_metric("accuracy", acc)
        mlflow.log_metric("roc_auc", roc_auc)
        
        # Log the confusion matrix and classification report
        mlflow.log_text(str(conf_matrix), "confusion_matrix.txt")
        mlflow.log_text(classification_report(y_test, y_pred), "classification_report.txt")
        
        # Log the model with signature and input example
        mlflow.sklearn.log_model(model, "model", signature=model_signature, input_example=input_example)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM, Conv1D, MaxPooling1D, Flatten

# LSTM Example
def build_lstm_model(input_shape):
    model = Sequential()
    model.add(LSTM(128, input_shape=input_shape, return_sequences=True))
    model.add(LSTM(64))
    model.add(Dense(1, activation='sigmoid'))  # Binary classification
    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
    return model

# Convert the data into time-series format for LSTM
X_train_reshaped = X_train.reshape((X_train.shape[0], 1, X_train.shape[1]))
X_test_reshaped = X_test.reshape((X_test.shape[0], 1, X_test.shape[1]))

lstm_model = build_lstm_model(input_shape=(1, X_train.shape[1]))

# Train LSTM
lstm_model.fit(X_train_reshaped, y_train, epochs=10, batch_size=64, validation_split=0.2)

# Evaluate LSTM
lstm_model.evaluate(X_test_reshaped, y_test)
```
